```python
# sample text
text="Artificial intelligence is changing the world . google is located in USA"
```

```python
from nltk.tokenize import sent_tokenize
sentences = sent_tokenize(text)
print("Segmentation:",sentences)
```

```python
from nltk.tokenize import word_tokenize
tokens = word_tokenize(text)
print("Tokenization", tokens)
```

```python
from nltk.stem import PorterStemmer 
ps=PorterStemmer()
stemmed=[ps.stem(w) for w in tokens]
print("stemming", stemmed)
```

```python
from nltk.stem import WordNetLemmatizer 
lemm = WordNetLemmatizer()
lemmatized=[lemm.lemmatize(w) for w in tokens]
print("Lemmatization", lemmatized)
```

```python
pos_tags = nltk.pos_tag(tokens)
print("POS Tagging:", pos_tags)
```

```python
named_entities=nltk.ne_chunk(pos_tags)
print("Named.Entities:", named_entities)
```
